```python
df["Категория зарплаты"] = ["Низкая" if x < 60000 else "Высокая" for x in df["Зарплата"]]
```

```python
df.loc[20] = [21,"Антон", 32, "Разработчик", "IT", 85000, 6, "Высокая"]
```

```python
new = pd.DataFrame([
    [22,"Максим", 29, "Аналитик", "Аналитика", 70000, 4, "Высокая"],
    [23,"Ксения", 34, "Менеджер", "Продажи", 95000, 9, "Высокая"]
], columns=df.columns)

df = pd.concat([df, new], ignore_index=True)
```

```python
print(df)
```
