Написать программу, выполняющую следующие действия: ввод с клавиатуры данных в список; ввод с клавиатуры данных и добавление строк в DataFrame; записи должны быть упорядочены по датам рождения; вывод на экран информации о человеке, чья фамилия введена с клавиатуры; если такого нет, выдать на дисплей соответствующее сообщение. Дополнительно реализовать сохранение и загрузку данных в формате Parquet, а также удаление одной из колонок DataFrame. При работе с датой и временем использовать модуль datetime.

```python
import pandas as pd
from datetime import datetime

df = pd.DataFrame(columns=["last_name", "first_name", "zodiac", "birth_date"])
```

```python
def add_person():
    
    last_name = input("Фамилия: ")
    first_name = input("Имя: ")
    zodiac = input("Знак зодиака: ")
    day = int(input("День рождения: "))
    month = int(input("месяц (номер): "))
    year = int(input("год: "))
    
    birth_date = datetime(year, month, day)
    
    new_row = pd.DataFrame([{
        "last_name": last_name,
        "first_name": first_name,
        "zodiac": zodiac,
        "birth_date": birth_date
    }])
    global df
    df = pd.concat([df, new_row], ignore_index=True)
    print(f"Добавили: {last_name} {first_name}")
```

```python
def find():
    last_name = input("Enter last name: ")
    
    result = df[df["last_name"] == last_name]
    
    if len(result) == 0:
        print(f"Не найдено")
    else:
        print("\nFound:")
        for _, row in result.iterrows():
            birth = row["birth_date"].strftime("%d.%m.%Y")
            print(f"   {row['last_name']} {row['first_name']}, {row['zodiac']}, {birth}")
```

```python
def show_sorted():

    if len(df) == 0:
        return
    
    df_sorted = df.sort_values("birth_date")
    
    for _, row in df_sorted.iterrows():
        birth = row["birth_date"].strftime("%d.%m.%Y")
        print(f"   {row['last_name']} {row['first_name']} - {birth} ({row['zodiac']})")
```
